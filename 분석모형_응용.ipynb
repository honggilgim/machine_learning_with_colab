```python
dataset = [['Milk','Cookie','Apple','Beans','Eggs','Yogurt'],
           ['Coke','Cookie','Apple','Beans','Eggs','Yogurt'],
           ['Milk','Apple','Kidney Beans','Eggs'],
           ['Milk','Orange','Corn','Beans','Yogurt'],
           ['Corn','Cookie','Cookie','Beans','Ice cream','Eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python

import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = apriori(df, min_support = 0.6, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x)) # 길이 변수 생성
frequent_itemsets
```

길이 생성 이유는, 이런 조건문에 사용하기 위해서.
조건문은, 길이가 2 이상이고 서포트가 0.6 이상인 것을 뽑아냄.

```python
frequent_itemsets[(frequent_itemsets['length'] == 2) & (frequent_itemsets['support'] >= 0.6)]
```

association_rules 함수 사용 : 데이터프레임을 입력, 기본값, 향상도 등을 조건으로 사용, 각 메트릭스 별 최솟값을 지정하고 기본값 지정. Aprori 결과에서 패턴 발견, 이 때 최소 신뢰도는 0.7

```python
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
rules
```

```python
rules2 = association_rules(frequent_itemsets, metric = "lift",min_threshold = 1.2)
rules2
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x : len(x))
```

새로운 칼럼으로 갯수를 추가하고, 그것을 이용하여 조건문을 사용.

```python
rules[(rules['antecedent_len'] >= 2) & (rules['confidence'] > 0.75) & (rules['lift'] > 1.2)]
```
