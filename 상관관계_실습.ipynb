상관관계 : 통계량. 두 집단 사이의 관계를 파악.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.show
```

```python
x = range(0,100)
# 리스트 x의 값을 하나씩 꺼내서 제곱한 값을 리스트 y에 할당. 간단한 for문
y = [v * v for v in x]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, 'ro') 
#ro : 산점도에 그려질 각 점의 색상과 모양. r은 빨간색, o는  출력될 점의 모양 다양한 형태 존재.
plt.show()
```

```python
# 상관분석 실습

import numpy as np
np.random.seed(1)
x = np.random.randint(0,50,500) # 0 ~ 50 사이의 난수 500개 생성
y = x + np.random.normal(0,10,500) # 노이즈를 포함하는 x와 양의 상관관계를 가지는 y 생성

np.corrcoef(x,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
# 음의 상관관계

x = np.random.randint(0,50,500)
y = 100 - x + np.random.normal(0,5,500) # x와 음의 상관관계를 가지는 y 생성 노이즈도 포함됨

np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
plt.show()
# 반대 방향으로 된 것을 볼 수 있음.
```
